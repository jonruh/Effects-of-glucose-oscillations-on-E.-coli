```python
import os
import pandas as pd
```

```python
# define file name of the export file
file_name = "DDB_SD_FeatureCounts.tsv"
```

```python
DF_Counts=pd.DataFrame()

os.chdir("../assembly_pipeline_results_and_expression_qc/DDB_SD_#1/featureCounts/")

for file in os.listdir():
    if file.endswith('cds.txt'):
        sample=file[:-8]

        DF_Counts[sample]=None

        infile=open(file,'r')

        for line in infile:

            if line.startswith('b'):
                gene=line.split('\t')[0]

                count=line.split('\t')[-1][:-1]

                DF_Counts.loc[gene,sample]=count

        infile.close()
        
os.chdir("../../../../data/feature_counts/")

DF_Counts.to_csv(file_name,  sep="\t")
```
